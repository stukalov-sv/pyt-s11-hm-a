**Функция f(x) =(sin(x)/2 + cos(x)/2)2 * sin(x)**

```python
from sympy import *
x, y, z, u = symbols('x, y, z, u')
y = (sin(x)/2 + cos(x)/2) ** 2 * sin(x)
```

**Определить корни:**

```python
y_roots = solve(y)
print(y_roots)
root_ev = [i.evalf(n=3) for i in y_roots]
print(root_ev)
```

**Найти интервалы, на которых функция возрастает:**

```python
graf = plot(y, (x, -pi/4, 7*pi/4), ylabel='y(x) in (-pi/4, 7*pi/4)', show=True)
graf
z = diff(y)
# plot(z, (x, -pi/4, 7*pi/4), ylabel="y'(x) in (-pi/4, 7*pi/4)", show=True)
z_roots = solve(z)
z_ordered = sorted(z_roots)
z_ordered.append(z_ordered[0] + 2 * pi)
z_ordered.append(z_ordered[1] + 2 * pi)
z_result = [z_ordered[i] + z_ordered[i+1]-z_ordered[i] for i in range(len(z_ordered) - 1)]
increases = ()
for i in range(1, len(z_result)-1, 2):
    increases = increases + ((z_result[i].evalf(n=3), (z_result[i+1].evalf(n=3))),)                         
print(f'Functon increasing in intervals: {increases} with period 2*pi')
```

**Найти интервалы, на которых функция убывает:**

```python
graf = plot(y, (x, -pi/4, 7*pi/4), ylabel='y(x) in (-pi/4, 7*pi/4)', show=True)
graf
z = diff(y)
# plot(z, (x, -pi/4, 7*pi/4), ylabel="y'(x) in (-pi/4, 7*pi/4)", show=True)
z_roots = solve(z)
z_ordered = sorted(z_roots)
z_ordered.append(z_ordered[0] + 2 * pi)
z_ordered.append(z_ordered[1] + 2 * pi)
z_result = [z_ordered[i] + z_ordered[i+1]-z_ordered[i] for i in range(len(z_ordered) - 1)]
increases = ()
for i in range(0, len(z_result)-1, 2):
    increases = increases + ((z_result[i].evalf(n=3), (z_result[i+1].evalf(n=3))),)                         
print(f'Functon decreasing in intervals: {increases} with period 2*pi')
```

**Вычислить вершины:**

```python
function = lambdify(x, y)
vertices_list = [function(float(z_ordered[i])).round(13) for i in range(len(z_ordered) - 2)]
print(f'Functon vertices in interval (-7*pi/4, 7*pi/4): {vertices_list} repeat with period 2*pi')
```

**Определить промежутки, на котором f > 0:**

```python
graf = plot(y, (x, -pi, pi), ylabel='y(x) in (-pi, pi)', show=True)
graf
u = diff(diff(y))
u_function = lambdify(x, u)
u_list = [u_function(float(z_ordered[i])).round(1) for i in range(len(z_ordered) - 2) if function(float(z_ordered[i])).round(13) != 0]
ordered_root = sorted(root_ev)
inter_list = ((ordered_root[-1].evalf(n=3) * -1, (ordered_root[0].evalf(n=3))),)
for i in range(0, len(ordered_root)-1):
    inter_list = inter_list + ((ordered_root[i].evalf(n=3), (ordered_root[i+1].evalf(n=3))),)

result_list = [inter_list[i] for i in range(len(u_list)) if u_list[i] < 0]
print(f'Functon > 0 in intervals: {result_list} with period 2*pi')
```
